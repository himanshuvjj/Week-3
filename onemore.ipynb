Data Visualization Using Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.head()
```

Seaborn Line plot

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

Seaborn Scatter Plot

```python
import pandas as pd
import seaborn as sns
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=iris,hue="species",style="species")
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=iris,hue="petal_length")
```

Seaborn Histogram plot

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
Diamonds=sns.load_dataset("diamonds")
```

```python
Diamonds.head()
```

```python
sns.histplot(Diamonds['price'])
plt.show()
```

```python
sns.distplot(Diamonds['price'])
plt.show()
```

```python
sns.distplot(Diamonds['price'],kde=False,bins=15)
plt.show()
```

```python
sns.distplot(Diamonds['price'],color="orange",bins=15,vertical=True)
plt.show()
```

Joint Plot

```python
import pandas as pd
import seaborn as sns
iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.jointplot(x="sepal_length",y="petal_length",data=iris)
```

```python
sns.jointplot(x="sepal_length",y="petal_length",data=iris,color="green")
```

```python
sns.jointplot(x="sepal_length",y="petal_length",data=iris,color="green",kind="reg")
```

```python
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species")
plt.show()
```
