Data Visualization using matplotlib



```python
import numpy as np
from matplotlib import pyplot as plt
```

Line Plot of single line

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*3
```

```python
y
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth='3')
plt.title("Line Plot of single line")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

Line plot of two line 

```python
X=np.arange(1,11)
Y1=X*2
Y2=X*3
```

```python
plt.plot(X,Y1,color='g',linestyle=':',linewidth='2')
plt.plot(X,Y2,color='r',linestyle='-',linewidth='3')
plt.title("Line Plot of two lines")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

Bar Plot

```python
Student={'bob':31,'sam':45,'raj':65,'himanshu':89}
```

```python
keys=list(Student.keys())
values=list(Student.values())
```

```python
plt.bar(keys,values)
plt.title("marks distribution of students")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(keys,values,color='g')
plt.title("marks distribution of students")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

Scatter Plot

```python
x=[11,24,35,47,28,33,27,19]
```

```python
y1=[4,5,7,8,9,4,5,7]
y2=[14,22,25,17,30,44,53,23]
```

```python
plt.scatter(x,y1,marker='*',color='g')
plt.scatter(x,y2,marker='.',color='r')
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y1,marker='*',color='g',s=100)
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.subplot(1,2,2)
plt.scatter(x,y2,marker='.',color='r',s=150)
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

BOX PLOT

```python
one=[1,2,3,4,5,5,6,7,8]
two=[3,4,6,2,6,8,6,5,4]
three=[4,6,8,9,5,7,8,9,5]
```

```python
data=list([one,two,three])
```

```python
plt.boxplot(data)
plt.show()
```

VIOLIN PLOT

```python
one=[1,2,3,4,5,5,6,7,8]
two=[3,4,6,2,6,8,6,5,4]
three=[4,6,8,9,5,7,8,9,5]
```

PIE CHART

```python
student=['gullu','yogesh','raj','sam']
marks=[20,44,3,34]
```

```python
plt.pie(marks,labels=student,autopct='%0.1f%%',colors=['white','blue','green','red'])
plt.show()
```

```python
student=['','yogesh','raj','sam']
marks=[20,474,36,304]
plt.pie(marks,labels=student,colors=['white','blue','green','red'])
plt.show()
```

DONAT CHART

```python
student=['vijay','yogesh','raj','sam']
marks=[20,64,36,74]
plt.pie(marks,labels=student,colors=['grey','blue','green','red'],radius=2)
plt.pie([1],radius=1,colors='w')
plt.show()
```
